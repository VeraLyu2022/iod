```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [119, 120, 121, 122, 123, 124, 125, 126],  # Number of trees in the forest
    'max_depth': [6, 7, 8, 9, 10, 11, 12],  # Maximum depth of the tree
    'min_samples_split': [1, 2, 3, 4, 5, 6, 7],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [5, 6, 7, 8, 9, 10, 11],  # Minimum number of samples required to be at a leaf node
}
clf = RandomForestClassifier()
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, 
                           cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.get_params()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

def model_summary_report(classifier, X, y):
    # Predictions
    y_pred = classifier.predict(X)
    
    # Accuracy
    accuracy = accuracy_score(y, y_pred)
    print(f"Accuracy: {accuracy:.4f}\n")
    
    
    # Classification Report
    report = classification_report(y, y_pred)
    print("Classification Report:\n", report)
model_summary_report(best_model, X_test, y_test)
```

```python
model_summary_report(best_model, X_train, y_train)
```

```python
y_train_proba = best_model.predict_proba(X_train)[:, 1]
show_summary_report(y_train, y_train_proba)
```
