## Lab 4.2.1: Feature Selection

### 1. Load & Explore Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### 1.1 Load Data

```python
# Read CSV
wine_csv = '../../DATA/winequality_merged.csv'
wine = pd.read_csv(wine_csv)
wine.head()
```

#### 1.2 Explore Data (Exploratory Data Analysis)

```python
# ANSWER
wine.corr()
```

```python
# using normalized data to calculate the correlation reduces the bias of the scales. 
# it didn't show much difference tho before normalization and after. 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(wine)
normalized_df = pd.DataFrame(normalized_data, columns=wine.columns)
normalized_df.corr()
```

### 2. Set Target Variable

Create a target variable for wine quality.

```python
# Target Variable
y = wine['quality']
y
```

### 3. Set Predictor Variables

Create a predictor matrix with variables of your choice. State your reasoning for the choices you make.

```python
# ANSWER
# choose the variables that has absolute correlation more than 10% to quality
# red wine column is a binary, it could be another target variable instead of predictor. 
X = wine[['volatile acidity', 'chlorides', 'density', 'alcohol']]
X
```

### 4. Using Linear Regression Create a Model and Test Score

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Create a model for Linear Regression
model = LinearRegression()

# Fit the model with the Training data
model_fit = model.fit(X_train, y_train)

# Calculate the score (R^2 for Regression) for Training Data
model_score_train = model.score(X_train, y_train)

# Calculate the score (R^2 for Regression) for Testing Data
model_score_test = model.score(X_test, y_test)

print(model_score_train, model_score_test)
```

## BONUS: Cross validation

```python
# Cross validation
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
# Set up 5-fold cross validation
k_fold = KFold(5, shuffle=True)
train_scores = []
train_rmse = []
test_scores = []
test_rmse = []

for k, (train, test) in enumerate(k_fold.split(X)):

    # Get training and test sets for X and y
    
    X_np = X.values
    y_np = y.values
    X_fold_train, X_fold_test = X_np[train], X_np[test]
    y_fold_train, y_fold_test = y_np[train], y_np[test]
    
    # Fit model with training set
    model_fit = model.fit(X_fold_train, y_fold_train)
    # Make predictions with training and test set
    predict_train = model_fit.predict(X_fold_train)
    predict_test = model_fit.predict(X_fold_test)
    # Score R2 and RMSE on training and test sets and store in list
    train_score = model_fit.score(X_fold_train, y_fold_train)
    train_scores.append(train_score)
    test_score = model_fit.score(X_fold_test, y_fold_test)
    test_scores.append(test_score)
    train_rmse_in = mean_squared_error(y_fold_train, predict_train, squared=False)
    train_rmse.append(train_rmse_in)
    test_rmse_in = mean_squared_error(y_fold_test, predict_test, squared=False)
    test_rmse.append(test_rmse_in)
# Create a metrics_df dataframe to display r2 and rmse scores
metrics_df = pd.DataFrame({'train_scores': train_scores, 'test_scores': test_scores, 'train_rmse': train_rmse, 'test_rmse': test_rmse})
metrics_df
```
